### The next cell is the most important cell

In this next cell, the code iterates over all the INMS text files contained in the directory. This will be read into a pandas dataframe, with which erranous lines are going to be removed. The dataframe is then going to be saved as a text file into a new directory. 

```python
import os 

### changes working directory to 
os.chdir(r"C:\Users\rifat\OneDrive - University College London\Rifat's PC\UCL\Y4\Project\MATLAB\Ali's MATLAB Work-20191010T175223Z-001\Ali_s MATLAB Work\Plots")
directory = r"C:\Users\rifat\OneDrive - University College London\Rifat's PC\UCL\Y4\Project\MATLAB\Ali's MATLAB Work-20191010T175223Z-001\Ali_s MATLAB Work\Plots"

### loops over all the files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        data = pd.read_csv(filename, sep = ' ', header = None, error_bad_lines = False)
        ### new directory to save new corrected files
        directory = r"C:\Users\rifat\OneDrive - University College London\Rifat's PC\UCL\Y4\Project\MATLAB\CorrectINMSdata"
        print(os.path.join(directory,filename))
        data.to_csv(os.path.join(directory,filename), header=None, index=None, sep=' ', mode='a')
    else:
        continue

```
